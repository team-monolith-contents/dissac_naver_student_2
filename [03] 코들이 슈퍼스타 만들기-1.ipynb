‘만일~이라면’ 부분은 앞부분과 똑같이 if문을 작성하고, ‘아니면’ 부분에는 else문을 작성합니다.

> **if-elif-else문**  
> - `if문`: 만약 조건이 참(True)이면 if 블록 안에 있는 코드를 실행하고 빠져나갑니다.
> - `elif문`: "else if"의 줄임말입니다. if가 거짓(False)이면, elif 조건을 검사합니다. 만약 elif 조건이 참이라면 해당 블록 안에 있는 코드를 실행합니다.
> - `else문`: 모든 if나 elif 조건이 거짓일 때 실행되는 부분입니다.
